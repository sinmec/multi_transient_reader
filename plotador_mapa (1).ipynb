```python
import pandas as pd

nome_arquivo_excel =Path(r'C:\Users\Gustavo\Documentos\Engenharia Mecânica - UFSC\Iniciação Cientica - SINMEC\ALFASIM\rodadas_parametricas.xlsx')
df = pd.read_excel(nome_arquivo_excel,skiprows=1)

parametric_runs_relation = {}

for index, row in df.iterrows():

    posicao = row[df.columns[0]]
    valor_c = row[df.columns[2]]
    valor_e = row[df.columns[4]]

    if valor_c >= 1:
        valor_c = round(valor_c)
    elif valor_c < 1 and valor_c >= 0.1:
        valor_c = round(valor_c, 1)
    elif valor_c < 0.1 and valor_c >= 0.01:
        valor_c = round(valor_c, 2)

    if valor_e >= 1:
        valor_e = round(valor_e)
    elif valor_e < 1 and valor_e >= 0.1:
        valor_e = round(valor_e, 1)
    elif valor_e < 0.1 and valor_e >= 0.01:
        valor_e = round(valor_e, 2)

    if posicao == "Base Run":
        posicao = "0"
    parametric_runs_relation[tuple([valor_c,valor_e])] = str(posicao)
```

```python
lista_01_a_01 = [round(valor, 2) for valor in np.arange(0.01, 0.11, 0.01)]
lista_1_a_10 = [round(valor, 1) for valor in np.arange(0.1, 1.1, 0.1)]
lista_10_a_100 = [round(valor) for valor in np.arange(1, 11, 1)]
```

```python
M_G_values = np.array(lista_01_a_01+lista_1_a_10+lista_10_a_100)
M_L_values = np.array(lista_01_a_01+lista_1_a_10+lista_10_a_100)
```

```python
M_L_grid, M_G_grid = np.meshgrid(M_G_values, M_L_values)
FP_grid_mode = np.full_like(M_G_grid, fill_value=0.0)
FP_grid_median = np.full_like(M_G_grid, fill_value=0.0)
```

```python
for i in range(len(M_L_values)):
    for j in range(len(M_G_values)):
        case = parametric_runs_relation[(M_G_values[j], M_L_values[i])]
        if case not in alfasim_result_1:
            FP_grid_median[i, j] = np.nan
            FP_grid_mode[i, j] = np.nan
            continue
        else:
            for edge in alfasim_result_1[case]:
                if edge == 'time':
                    time = alfasim_result_1[case]['time']
                    continue
                FP = alfasim_result_1[case][edge]["14.326"]['Flow Pattern']['values']
                FP_nan = FP[~np.isnan(FP)]
                if case <= '350':
                    avg_FP = int(np.nanmedian(FP_nan[-240:]))
                    mode_FP = int(statistics.mode(FP_nan[-240:]))
                else:
                    avg_FP = int(np.nanmedian(FP_nan[-24:]))
                    mode_FP = int(statistics.mode(FP_nan[-24:]))
                FP_grid_median[i, j] = avg_FP
                FP_grid_mode[i, j] = mode_FP
```

```python
plt.pcolormesh(M_L_grid, M_G_grid, FP_grid_mode)
plt.title("Mode")
plt.colorbar()
plt.xscale('log')
plt.yscale('log')
plt.xlim(0.01, 10)
plt.ylim(0.01, 10)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color="black")
plt.xlabel('j_g [m/s]')
plt.ylabel('j_l [m/s]')
plt.grid()
plt.show()
```

```python
plt.pcolormesh(M_L_grid, M_G_grid, FP_grid_median)
plt.title("Median")
plt.colorbar()
plt.xscale('log')
plt.yscale('log')
plt.xlim(0.01, 10)
plt.ylim(0.01, 10)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color="black")
plt.xlabel('j_g [m/s]')
plt.ylabel('j_l [m/s]')
plt.grid()
plt.show()
```
